```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyser = SentimentIntensityAnalyzer()
```

```python
import pandas as pd
```

```python
file_name = pd.read_excel("C:/Users/SUMESH.IYER/Desktop/data.xlsx")
file_name
```

```python
def sentiment_analyzer_scores(sentence):
    score = analyser.polarity_scores(sentence)
#     print("{:-<40} {}".format(sentence, str(score)))
    return (score)
```

```python
sentiment = []
temp = []
file_name.fillna({'Translated_Review':0}, inplace=True)
for i in file_name.Translated_Review:
    if(i==0):
        sentiment.append("NA")
    else:
        score_v = sentiment_analyzer_scores(i)
        print(i)
        del score_v['compound']
#         print(max(score_v.values()))
#         print(score_v.items())
        temp = sorted(score_v.items(), key=lambda kv: kv[1])
        temp = temp[2]
        sentiment.append(temp[0])
print(sentiment)
    
    
        
        
```
